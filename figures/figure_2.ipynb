```python
import seaborn as sns
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np

import warnings
warnings.filterwarnings('ignore')

from matplotlib import rcParams

sns.set_style('whitegrid', {'xtick.color': '.15',
 'xtick.direction': u'in',
 'xtick.major.size': 10,
 'xtick.minor.size': 5,
 'ytick.color': '.15',
 'ytick.direction': u'in',
 'ytick.major.size': 10,
 'ytick.minor.size': 5,
 'grid.color': '.8',
 'axes.edgecolor': '.4',
 'grid.linestyle': u'--',
 'axes.linewidth': 1.0,
 'grid.linewidth': 0.5})

SMALL_SIZE = 20
MEDIUM_SIZE = 20
BIGGER_SIZE = 20

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

paired = sns.color_palette("Paired", 10)
```

```python
f = plt.figure()
f.set_size_inches(19, 10)
plt.rc('axes', axisbelow=False)

plt.subplots_adjust(hspace=0.001)
plt.subplots_adjust(wspace=0.15)

dat = sio.loadmat('../data/figure_2/ice_shelf_draft_melt_fig2')
draft = np.squeeze(dat['draft'])
N = dat['N'] 
d_draft = 30.
gridsize = 500.; rho_i = 917.; fac = (gridsize**2)*rho_i/1e12
melt = np.multiply(fac*dat['melt'],N)/d_draft
melt_std = np.multiply(fac*dat['melt_std'],N)/d_draft
steady = np.multiply(fac*dat['steady'],N)/d_draft
                   
ax1 = plt.subplot2grid((1,6),(0,0),rowspan=1,colspan=6)

i = 22 #Getz
off = 1.95
ax1.axvline(off,color='black',alpha=0.1)
wbf = np.squeeze(melt[i,:]) + off; wbf_std = np.squeeze(melt_std[i,:]); 
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)<=0,facecolor=paired[0],interpolate=False)
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)>=0,facecolor=paired[4],interpolate=False)
ax1.plot(wbf + wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--')
ax1.plot(wbf - wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--',label='95% confidence \nintervals')
sf = np.squeeze(steady[i,:]) + off
ax1.plot(sf[:-57],draft[:-57],color=paired[9],linestyle='-',label='steady-state melting',linewidth=2)
ax1.text(off,1230,'Getz',horizontalalignment='center',fontsize=22)
ax1.set_ylim([0,2200])
ax1.legend(loc='best', bbox_to_anchor=(0.5, 0., 0.5, 0.195), framealpha=1)
ax1.set_xlim([0,2.25])
plt.ylabel("Depth ($m$)")

i = 28#Ross
off = 0.25
ax1.axvline(off,color='black',alpha=0.1)
wbf = np.squeeze(melt[i,:]) + off; wbf_std = np.squeeze(melt_std[i,:]); 
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)<=off,facecolor=paired[0],interpolate=True)
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)>=off,facecolor=paired[4],interpolate=True)
ax1.plot(wbf + wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--')
ax1.plot(wbf - wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--',label='95% confidence \nintervals')
sf = np.squeeze(steady[i,:]) + off
ax1.plot(sf,draft,color=paired[9],linestyle='-',label='steady-state melting',linewidth=2)
ax1.text(off,1500,'Ross',horizontalalignment='center',fontsize=22)
plt.ylabel("Depth ($m$)")

i = 36 #Totten
off = 1.75
ax1.axvline(off,color='black',alpha=0.1)
wbf = np.squeeze(melt[i,:]) + off; wbf_std = np.squeeze(melt_std[i,:]); 
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)<=off,facecolor=paired[0],interpolate=True)
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)>=off,facecolor=paired[4],interpolate=True)
ax1.plot(wbf + wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--')
ax1.plot(wbf - wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--',label='95% confidence \nintervals')
sf = np.squeeze(steady[i,:]) + off
ax1.plot(sf,draft,color=paired[9],linestyle='-',label='steady-state melting',linewidth=2)
t = ax1.text(off,2130,'Totten',horizontalalignment='center',backgroundcolor='white',fontsize=22)
t.set_bbox(dict(facecolor='white', alpha=0.8, edgecolor='white'))
plt.ylabel("Depth ($m$)")

i = 40 #Amery
off = 1.5
ax1.axvline(off,color='black',alpha=0.1)
wbf = np.squeeze(melt[i,:]) + off; wbf_std = np.squeeze(melt_std[i,:]); 
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)<=off,facecolor=paired[0],interpolate=True)
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)>=off,facecolor=paired[4],interpolate=True)
ax1.plot(wbf + wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--')
ax1.plot(wbf - wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--',label='95% confidence \nintervals')
sf = np.squeeze(steady[i,:]) + off
ax1.plot(sf,draft,color=paired[9],linestyle='-',label='steady-state melting',linewidth=2)
t = ax1.text(off,2130,'Amery',horizontalalignment='center',backgroundcolor='white',fontsize=22)
t.set_bbox(dict(facecolor='white', alpha=0.8, edgecolor='white'))
plt.ylabel("Depth ($m$)")

i = 6 #Filchner
off = 1.25
ax1.axvline(off,color='black',alpha=0.1)
ax1.axvline(off,color='black',alpha=0.1)
wbf = np.squeeze(melt[i,:]) + off; wbf_std = np.squeeze(melt_std[i,:]); 
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)<=off,facecolor=paired[0],interpolate=True)
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)>=off,facecolor=paired[4],interpolate=True)
ax1.plot(wbf + wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--')
ax1.plot(wbf - wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--',label='95% confidence \nintervals')
sf = np.squeeze(steady[i,:]) + off
ax1.plot(sf,draft,color=paired[9],linestyle='-',label='steady-state melting',linewidth=2)
ax1.text(off,2130,'Filchner',horizontalalignment='center',fontsize=22)

i = 5 #Ronne
off = 0.85
ax1.axvline(off,color='black',alpha=0.1)
wbf = np.squeeze(melt[i,:]) + off; wbf_std = np.squeeze(melt_std[i,:]); 
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)<=off,facecolor=paired[0],interpolate=True)
ax1.fill_betweenx(draft,np.squeeze(wbf),x2=off,where= np.squeeze(wbf)>=off,facecolor=paired[4],interpolate=True)
ax1.plot(wbf + wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--')
ax1.plot(wbf - wbf_std,draft,color=(0.5,0.5,0.5),linestyle='--',label='95% confidence \nintervals')
sf = np.squeeze(steady[i,:]) + off
ax1.plot(sf,draft,color=paired[9],linestyle='-',label='steady-state melting',linewidth=2)
ax1.text(off,2130,'Ronne',horizontalalignment='center',fontsize=22)
plt.ylabel("Ice draft (m)")
ax1.set_xticks([0.25,0.5,0.75,1,1.25,1.5,1.75,2])
plt.xlim([0.05,2.45])

off = 0.55
ax1.text(off,1650,'Melting',horizontalalignment='center',fontsize=22,color=paired[5])
ax1.text(off,1750,'Freezing',horizontalalignment='center',fontsize=22,color=paired[1])

plt.plot([off-0.1,off+0.1], [1390, 1390], color='k', linestyle='-', linewidth=4)
ax1.text(off,1360,'0.2 Gt/(m$\cdot$yr)',fontsize=22,horizontalalignment='center')

plt.minorticks_on()
plt.grid(b=True, which='minor', color='k', linestyle='-', alpha=0)
plt.grid(b=True, which='major', color='k', linestyle='-', alpha=0)

ax1.set_ylim(ax1.get_ylim()[::-1])

plt.xlabel("Area-integrated meltwater flux (20102018) per m of ice draft")

plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
```
