```python
import seaborn as sns
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import rcParams

sns.set_style('whitegrid', {'xtick.color': '.15',
 'xtick.direction': u'in',
 'xtick.major.size': 10,
 'xtick.minor.size': 5,
 'ytick.color': '.15',
 'ytick.direction': u'in',
 'ytick.major.size': 10,
 'ytick.minor.size': 5,
 'grid.color': '.8',
 'axes.edgecolor': '.4',
 'grid.linestyle': u'--',
 'axes.linewidth': 1.0,
 'grid.linewidth': 0.5})

SMALL_SIZE = 20
MEDIUM_SIZE = 20
BIGGER_SIZE = 20

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=BIGGER_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

paired = sns.color_palette("Paired", 10)
```

```python
f = plt.figure()
f.set_size_inches(9, 18)
plt.rc('axes', axisbelow=False)

plt.subplots_adjust(hspace=0.1)
plt.subplots_adjust(wspace=0)

dat = sio.loadmat('../data/figure_4/deep_shallow_melting.mat')
time = dat['t']

ax1 = plt.subplot2grid((5,1),(3,0),rowspan=1,colspan=1)
ax1.axhline(y=0,color='black',alpha=0.3)
wb = dat['wb_amery_deep']
wb_std = dat['wb_amery_deep_std']
ax1.plot(time,wb,color=paired[5],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[4], facecolor=paired[4], linewidth=0)
wb = dat['wb_amery_front']
wb_std = dat['wb_amery_front_std']
ax1.plot(time,wb,color=paired[1],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[0], facecolor=paired[0], linewidth=0)
ax1.set_xlim(1994.1,2019)
ax1.set_xticks([1996,2000,2004,2008,2012,2016])
ax1.set_yticks([0,4,8])
ax1.set_ylim(-2,8)
plt.minorticks_on()
plt.grid(b=True, which='minor', color='k', linestyle='-', alpha=0.1)
plt.grid(b=True, which='major', color='k', linestyle='-', alpha=0.2)
ax1.tick_params(labelbottom=False)    

ax1 = plt.subplot2grid((5,1),(1,0),rowspan=1,colspan=1)
ax1.axhline(y=0,color='black',alpha=0.3)
wb = dat['wb_ronne_deep']
wb_std = dat['wb_ronne_deep_std']
ax1.plot(time,wb,color=paired[5],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[4], facecolor=paired[4], linewidth=0)
wb = dat['wb_ronne_front']
wb_std = dat['wb_ronne_front_std']
ax1.plot(time,wb,color=paired[1],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[0], facecolor=paired[0], linewidth=0)
ax1.set_xlim(1994.1,2019)
ax1.set_xticks([1996,2000,2004,2008,2012,2016])
ax1.set_ylim(-0.5,3)
ax1.set_yticks([0,1,2,3])
plt.minorticks_on()
plt.grid(b=True, which='minor', color='k', linestyle='-', alpha=0.1)
plt.grid(b=True, which='major', color='k', linestyle='-', alpha=0.2)
ax1.tick_params(labelbottom=False)  

ax1 = plt.subplot2grid((5,1),(2,0),rowspan=1,colspan=1)
ax1.axhline(y=0,color='black',alpha=0.3)
wb = dat['wb_filchner_deep']
wb_std = dat['wb_filchner_deep_std']
ax1.plot(time,wb,color=paired[5],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[4], facecolor=paired[4], linewidth=0)
wb = dat['wb_filchner_front']
wb_std = dat['wb_filchner_front_std']
ax1.plot(time,wb,color=paired[1],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[0], facecolor=paired[0], linewidth=0)
ax1.set_xlim(1994.1,2019)
ax1.set_xticks([1996,2000,2004,2008,2012,2016])
ax1.set_ylim(-1,2)
ax1.set_yticks([0,1,2])
plt.minorticks_on()
plt.grid(b=True, which='minor', color='k', linestyle='-', alpha=0.1)
plt.grid(b=True, which='major', color='k', linestyle='-', alpha=0.2)
ax1.tick_params(labelbottom=False)    
plt.ylabel('Basal melt rate (m/yr)')  

ax1 = plt.subplot2grid((5,1),(0,0),rowspan=1,colspan=1)
wb = dat['wb_ross_deep']
wb_std = dat['wb_ross_deep_std']
ax1.plot(time,wb,color=paired[5],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[4], facecolor=paired[4], linewidth=0)
wb = dat['wb_ross_front']
wb_std = dat['wb_ross_front_std']
ax1.plot(time,wb,color=paired[1],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[0], facecolor=paired[0], linewidth=0)
ax1.set_xlim(1994.1,2019)
ax1.set_xticks([1996,2000,2004,2008,2012,2016])
ax1.set_ylim(0,8)
ax1.tick_params(labelbottom=False)    
plt.minorticks_on()
plt.grid(b=True, which='minor', color='k', linestyle='-', alpha=0.1)
plt.grid(b=True, which='major', color='k', linestyle='-', alpha=0.2)
ax1.set_yticks([2,4,6,8])

ax1 = plt.subplot2grid((5,1),(4,0),rowspan=1,colspan=1)
wb = np.squeeze(dat['wb_ra_amundsen'])
wb_std = np.squeeze(dat['wb_ra_amundsen_std'])
ax1.plot(time,wb,color=paired[9],marker='.',markersize=0.1)
ax1.fill_between(np.array(np.squeeze(time)), np.array(np.squeeze(wb+wb_std)), np.array(np.squeeze(wb-wb_std)),
    alpha=0.3, edgecolor=paired[8], facecolor=paired[8], linewidth=0)
ax1.set_xlim(1994.1,2019)
ax1.set_yticks([5,6,7,8,9])
ax1.set_ylim(5,9)
ax1.set_xticks([1996,2000,2004,2008,2012,2016])
plt.minorticks_on()
plt.grid(b=True, which='minor', color='k', linestyle='-', alpha=0.1)
plt.grid(b=True, which='major', color='k', linestyle='-', alpha=0.2)

plt.show()
```
